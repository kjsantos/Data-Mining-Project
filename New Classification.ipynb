```python
from matplotlib import pyplot as plt
import matplotlib as mp
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import sklearn.manifold
from scipy import cluster
import seaborn as sns
import sklearn.datasets as sk_data
import sklearn.metrics as metrics
import requests
import json
import nltk.classify.util
from nltk.classify import NaiveBayesClassifier
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk import word_tokenize
import re
import json
import pickle
%matplotlib inline
```

```python
try:
    with open('df_fi_sorted','rb') as f:
        df_fi_sorted = pickle.load(f)
        
except FileNotFoundError:
    with open('df_fi_sorted','wb') as f:
        pickle.dump(df_fi_sorted, f)
```

```python
df_fi_sorted
```

```python
file = 'yelp_academic_dataset_business.json'
closed=[]
stars=[]
long=[]
lat=[]
attributes = []
with open(file) as f:
    for line in f:
        r_line = json.loads(line)
        if r_line['business_id'] in set(df_fi_sorted['business_id']):
            closed.append(r_line['open'])
            stars.append(r_line['stars'])
            long.append(r_line['longitude'])
            lat.append(r_line['latitude'])
            attributes.append(r_line['attributes'])
closed = np.asarray(closed)
stars = np.asarray(stars)
long = np.asarray(long)
lat = np.asarray(lat)
attributes
```

```python
price_range=[]
for x in attributes:
    found=False
    for x,y in x.items():
        #print(x)
        if x=='Price Range':
            found=True
            price_range.append(y)
            continue
    if found==False:
        price_range.append('None')
price_range = np.asarray(price_range)
price_range
```

```python
df_fi_sorted['open']=closed
df_fi_sorted['stars']=stars
df_fi_sorted['long']=long
df_fi_sorted['lat']=lat
df_fi_sorted['price_range']=price_range
```

```python
df_fi_sorted
```

```python
df_fi_classifier = df_fi_sorted.drop(['index'],axis=1)
df_fi_classifier = df_fi_classifier.drop(['text'],axis=1)
df_fi_classifier = df_fi_classifier.drop(['votes'],axis=1)
df_fi_classifier = df_fi_classifier.drop(['fi_scores'],axis=1)
df_fi_classifier = df_fi_classifier.drop(['business_id'],axis=1)
```

```python
df_fi_classifier
```

```python
df_star = pd.get_dummies(df_fi_sorted['stars'], prefix = 'stars')
df_open = pd.get_dummies(df_fi_sorted['open'], prefix = 'open')
df_price = pd.get_dummies(df_fi_sorted['price_range'], prefix = 'price')
df_fi_classifier = pd.concat([df_fi_classifier, df_star], axis=1)
df_fi_classifier = pd.concat([df_fi_classifier, df_open], axis=1)
df_fi_classifier = pd.concat([df_fi_classifier, df_price], axis=1)
df_fi_classifier = df_fi_classifier.drop(['stars'], axis = 1)
df_fi_classifier = df_fi_classifier.drop(['open'], axis = 1)
df_fi_classifier = df_fi_classifier.drop(['price_range'], axis = 1)
df_fi_classifier = df_fi_classifier.drop(['open_True'], axis = 1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.cross_validation import train_test_split

acc=[]
for i in range(1,60):
    X = df_fi_classifier.drop(['open_False'], axis=1)
    X.drop(['open_True'], axis=1)
    y = df_fi_classifier['open_False']
    X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=.75)
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred = knn.predict(X_test)
    #print(len(pred))
    test_score = knn.score(X_train,y_train)
    #print('Accuracy on test data: {}'.format(test_score))
    acc.append(test_score)
    #print('finished {} test: '.format(i))

```

```python
plt.scatter(list(range(2,60)),acc[1:],s=20)
```

```python
X = df_fi_classifier.drop(['open_False'], axis=1)
X.drop('open_True', axis=1)
y = df_fi_classifier['open_False']
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=.7)
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train,y_train)
pred = knn.predict(X_test)
pred.reshape(y_test.shape)
print(pred)
test_score = knn.score(X_train,y_train)
print('Accuracy on test data: {}'.format(test_score))
```

```python
from matplotlib.colors import ListedColormap
train_guess=[]
test_guess=[]
for x in range(len(y_train)):
    if y_train.iloc[x,0]==1:
        train_guess.append(1)
    else:
        train_guess.append(0)
for x in range(len(y_test)):
    if y_test.iloc[x,0]==1:
        test_guess.append(1)
    else:
        test_guess.append(0)
        
cmap_light = ListedColormap(['#AAFFAA','#FFAAAA'])
cmap_bold = ListedColormap(['#00FF00','#FF0000'])
plt.subplot()
plt.scatter(X_train['lat'], X_train['long'], c=train_guess,cmap=cmap_light)
plt.axis('equal')
plt.title('Training using KNN')
plt.show()
plt.subplot()
plt.scatter(X_test['lat'], X_test['long'], c=test_guess,cmap=cmap_light)
plt.title('Testing $k$={}\nAccuracy: {}'.format(3,test_score))
_=plt.axis('equal')
```
